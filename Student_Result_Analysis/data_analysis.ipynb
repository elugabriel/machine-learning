```python
conditions = [
    (data['AVERAGE'] >= 70),
    (data['AVERAGE'] >= 60) & (data['AVERAGE'] <= 69),
    (data['AVERAGE'] >= 50) & (data['AVERAGE'] <= 59),
    (data['AVERAGE'] >= 40) & (data['AVERAGE'] <= 49),
    (data['AVERAGE'] < 40)
    ]


values = ['First Class', 'Second Class Upper', 'Secon Class Lower', 'Third Class','Fail']

data['CLASS'] = np.select(conditions, values)
```
